```python
import json
import random
```

Mounting the drive is necessary step to load the dataset:

```python
from google.colab import drive
drive.mount('/content/drive')
database_path = '/content/drive/MyDrive/Discord_Manager_files/discord_commands.jsonl'
```

Generating unique names and roles, we decided to cap it to 100.
we made templates that simulate how a user would probably use each action:

```python
# Generate unique names and roles
users = [f"<USER_{i}>" for i in range(1, 101)]  # 100 unique users
roles = [f"<ROLE_{i}>" for i in range(1, 100)]  # 99 unique roles, the
# ROLE_100> is saved for the role null as not all actions require a role

# Expanded action templates with placeholders
action_templates = {
    "ban": [
        "i think {user} needs to be banned",
        "ban {user} immediately",
        "{user} should be banned",
        "please ban {user}",
        "we need to ban {user}",
        "{user} is causing trouble, ban them",
        "ban {user} right now",
        "get rid of {user}, ban them",
        "{user} deserves to be banned",
        "ban {user} as soon as possible"
    ],
    "remove_role": [
        "please remove the {role} role from {user}",
        "remove {role} from {user}",
        "take away {role} from {user}",
        "revoke {role} from {user}",
        "{user} should no longer have the {role} role",
        "strip {user} of the {role} role",
        "remove the {role} role assigned to {user}",
        "{user} doesn't deserve the {role} role anymore",
        "take the {role} role away from {user}",
        "please revoke {role} from {user}"
    ],
    "add_role": [
        "add the {role} role to {user}",
        "grant {role} to {user}",
        "assign {role} to {user}",
        "give {user} the {role} role",
        "{user} should have the {role} role",
        "please assign {role} to {user}",
        "add {role} to {user}'s roles",
        "grant the {role} role to {user}",
        "give {user} access to the {role} role",
        "{user} deserves the {role} role"
    ]
}

```

we generate our dataset of 10,000 examples and save it to a json file, this will help us train the model.

```python
# Generate 10,000 examples
with open(database_path, "w") as f:
    for _ in range(10000):
        user = random.choice(users)
        action = random.choice(["ban", "remove_role", "add_role"])

        if action == "ban":
            template = random.choice(action_templates["ban"])
            input_text = template.format(user=user)
            output = {"action": "ban", "user": user}
        else:
            role = random.choice(roles)
            template = random.choice(action_templates[action])
            input_text = template.format(user=user, role=role)
            output = {"action": action, "user": user, "role": role}

        # Write each example as a single line
        f.write(json.dumps({"input": input_text, "output": output}) + "\n")

print("Dataset generated with 10,000 examples in JSON Lines format")
```
